```python
### nltk -> 필요한 기능을 추가로 다운로드

# 필요한 라이브러리 임폴트
import nltk

# 다운로드 함수 호출
nltk.download('all')
```

```python
# 필요한 함수 임폴트
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
### 문장 단위 토큰화

# text 데이터 생성
text_en ='God is Great! I won a lottery.'
text_ko = "얘들아 안녕! 나는 얼굴 천재 차은우야. 앞으로 잘 부탁해."

# 문장 단위 토큰화 -> sent_tokenize() 사용
tokenized_en = sent_tokenize(text=text_en)
tokenized_ko = sent_tokenize(text=text_ko)

# 결과 확인하기
print(f'영어 텍스트 데이터의 문장 단위 토큰화 결과 : \n{tokenized_en}')
print('-'*80)
print(f'한글 텍스트 데이터의 문장 단위 토큰화 결과 : \n{tokenized_ko}')
```

```python
### 단어 단위 토큰화

# text 데이터 생성
text_en ='God is Great! I won a lottery.'
text_ko = "얘들아 안녕! 나는 얼굴 천재 차은우야. 앞으로 잘 부탁해."

# 단어 단위 토큰화 -> word_tokenize()사용
tokenized_en = word_tokenize(text=text_en)
tokenized_ko = word_tokenize(text=text_ko)

# 문자열.split('')
splited_en = text_en.split(' ')
splited_ko = text_ko.split(' ')

# 결과 확인하기
print(f'영어 텍스트 데이터의 단어 단위 토큰화 한 결과 : \n{tokenized_en}')
print('-'*80)
print(f'영어 텍스트 데이터를 split()함수로 분할한 결과(느낌표까지 같이 출력) : \n{splited_en}')
print('-'*80)
print(f'한글 텍스트 데이터의 단어 단위 토큰화 한 결과 : \n{tokenized_ko}')
print('-'*80)
print(f'한글 텍스트 데이터를 split()함수로 분할한 결과 : \n{splited_ko}')
```

```python
### 필요한 함수 임폴트
from konlpy.tag import Okt
from mecab import MeCab
```

```python
### 한글 형태소 분석기 객체 생성
okt = Okt()
mecab = MeCab()
```

```python
### Okt 형태소 분석기 실습(1)

# 실습 문장 생성
text = '한글 형태소 분석기로 테스트를 해보았습니다. 설치와 동작이 잘 됩니다.'

# 문장 -> 형태소 단위로 분할(토큰화, 형태소 분석) -> okt.morphs(text)
morphemes_okt = okt.morphs(text)

# 비교 : word_tokenize()사용
result = word_tokenize(text)

# 결과 확인하기
print(f'Okt 형태소 분석기를 이용하여 형태소 분석(토큰화)를 한 결과 : \n{morphemes_okt}')
print('-'*80)
print(f'word_tokenize 함수를 이용하여 토큰화 한 결과 : \n{result}')
```

```python
### Mecab형태소 분석기 실습

# 실습 문장 생성
text = '한글 형태소 분석기로 테스트를 해보았습니다. 설치와 동작이 잘 됩니다.'

# 문장 -> 형태소 단위로 분할(토큰화, 형태소 분석) -> mecab.morphs(text)
morphs_mecab = mecab.morphs(text)

# 결과 확인하기
print(f'Mecab 형태소 분석기를 이용하여 형태소 분석(토큰화)를 한 결과 : \n{morphs_mecab}')
print('-'*80)
print(f'okt 함수를 이용하여 토큰화 한 결과 : \n{morphemes_okt}')
```

```python
### Okt 형태소 분석기 실습(2) - pos()

# 실습 문장 생성
text = '한글 형태소 분석기로 테스트를 해보았습니다. 설치와 동작이 잘 됩니다.'

# 문장 -> 형태소와 품사를 동시에 분석 및 추출 -> okt.pos(text)
pos_okt = okt.pos(text)

# 결과 확인하기
print(f'Okt 형태소 분석기를 이용하여 형태소와 품사를 동시에 추출한 결과 : \n{pos_okt}')
```

```python
### Mecab 형태소 분석기 실습(2) - pos()
#mecab은 일본어를 위해 개발이 되었기 때문에, pos 함수에 있어서는 mecab보다 okt가 더 정확하다.

# 실습 문장 생성
text = '한글 형태소 분석기로 테스트를 해보았습니다. 설치와 동작이 잘 됩니다.'

# 문장 -> 형태소와 품사를 동시에 분석 및 추출 -> mecab.pos(text)
pos_mecab = mecab.pos(text)

# 결과 확인하기
print(f'Mecab 형태소 분석기를 이용하여 형태소와 품사를 동시에 추출한 결과 : \n{pos_mecab}')
```

```python
### Okt 형태소 분석기 실습(3) - nouns()

# 실습 문장 생성
text = '한글 형태소 분석기로 테스트를 해보았습니다. 설치와 동작이 잘 됩니다.'

# 문장 --> 명사인 형태소 추출 --> okt.nouns(text)
nouns_okt = okt.nouns(text)

# 결과 확인하기
print(f'Okt 형태소 분석기를 이용하여 명사만 추출한 결과 : \n{nouns_okt}')
```

```python
### Mecab 형태소 분석기 실습(3) - nouns()

# 실습 문장 생성
text = '한글 형태소 분석기로 테스트를 해보았습니다. 설치와 동작이 잘 됩니다.'

# 문장 --> 명사인 형태소 추출 --> mecab.nouns(text)
nouns_mecab = mecab.nouns(text)

# 결과 확인하기
print(f'Mecab 형태소 분석기를 이용하여 명사만 추출한 결과 : \n{nouns_mecab}')
```

```python
### 한글 폰트 설치
!apt-get update -qq
!apt-get install fonts-nanum* -qq
```

```python
# 대한민국 헌법 텍스트 데이터 불러오기

# 파일 경로 설정
file_path = '/content/drive/MyDrive/KDT/딥러닝/자연어처리/constitution.txt'

# 파이썬으로 파일 읽어오기 (txt파일)
with open(file_path, 'r')as f:
    text = f.read()

# 결과 확인하기
print(f'대한민국 헌법 텍스트 데이터 : \n{text}')
print('-'*80)
print(f'대한민국 헌법 텍스트 데이터의 길이 = {len(text)}개')
```

```python
### Mecab 형태소 분석기 -> 명사 추출 -> mecab.nouns(text)
# 실제로는 okt보다 mecab을 주로 쓴다. 속도도 더 빠르며, 더 정확하다.

# Mecab 형태소 분석기 임폴트
from mecab import MeCab as Mecab

# mecab 형태소 분석기 생성 / 클래스를 생성해주면 3가지 함수를 사용 가능/ morphs, nouns, pos
mecab = Mecab()

# 형태소 분석 -> 명사 추출(nouns)
nouns = mecab.nouns(text)
print(f'mecab 형태소 분석기를 이용하여 명사만 추출한 결과 : \n{nouns}')
print(f'추출된 명사의 개수 : {len(nouns)}개')
```

```python
### Okt 형태소 분석기 -> 명사 추출 -> okt.nouns(text)

# Mecab 형태소 분석기 임폴트
from konlpy.tag import Okt

# mecab 형태소 분석기 생성 / 클래스를 생성해주면 3가지 함수를 사용 가능/ morphs, nouns, pos
okt = Okt()

# 형태소 분석 -> 명사 추출(nouns)
nouns_okt = okt.nouns(text)
print(f'Okt 형태소 분석기를 이용하여 명사만 추출한 결과 : \n{nouns_okt}')
print(f'추출된 명사의 개수 : {len(nouns_okt)}개')
```

```python
### '명사' 형태소의 빈도수 분석

# 필요한 함수 임폴트
from collections import Counter

# 명사의 빈도수 분석
counts = Counter(nouns)
print(f'명사의 빈도수 : {counts}')

print('-'*80)

# 딕셔너리의 하위 형태. 딕셔너리처럼 사용하면 된다.(<class 'collections.Counter'>)
print(type(counts))
```

```python
### 불용어 제거 -> 명사의 길이가 1보다 큰 명사만 추출

# 결과를 저장할 빈 리스트 생성
result = []

# for문 + if 조건문 실행
for noun in nouns:
    if len(noun) > 1:
        result.append(noun)

# 결과 확인하기
print(f'길이가 1보다 큰 명사인 형태소 확인 : \n{result}')
print('-'*80)
print(f'길이가 1보다 큰 명사인 형태소의 개수 : {len(result)}개')
```

```python
### 명사로 된 형태소의 빈도수 분석

# 명사인 형태소의 빈도수 분석
counts = Counter(result)

# 결과 확인하기
print(f'길이가 1보다 큰 명사의 빈도수 \n{counts}')
print('-'*80)
print(f'길이가 1보다 큰 명사의 종류 : {len(counts)}')
```
